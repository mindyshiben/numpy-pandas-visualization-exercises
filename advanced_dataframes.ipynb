### Part 2 Exercises

```python
# Copy the users and roles DataFrames from the examples above.

users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})

roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})

users, roles
```

```python
# What is the result of using a right join on the DataFrames?

users.merge(roles, left_on='role_id', right_on='id', how='right')

# includes all data from roles, even when there are null values for cooresponding records in users. 
# Users data that has cooresponding null values in roles are not returned
```

```python
# What is the result of using an outer join on the DataFrames?

users.merge(roles, left_on='role_id', right_on='id', how='outer')

# outer includes all data regardless of null values on either table
```

```python
# What happens if you drop the foreign keys from the DataFrames and try to merge them?

users.merge(roles,how='inner')
users.merge(roles,how='outer')

# nothing is returned with 'inner'
# with 'outer', all values are returned, but nothing is merged. 
# Keys to merge on need to be specified for the merge to be useful
```

```python
# How many different manufacturers are there?

len(mpg['manufacturer'].unique())  #15

# .nunique() works here too and will not have to use len -- like combining unique/len
```

```python
# How many different models are there?

len(mpg['model'].unique())  #38
```

```python
# Create a column named mileage_difference like you did in the DataFrames exercises; 
# this column should contain the difference between highway and city mileage for each car.

mpg['mileage_difference']= abs(mpg['hwy']-mpg['city'])
mpg
```

```python
# Create a column named average_mileage like you did in the DataFrames exercises; 
# this is the mean of the city and highway mileage.

mpg['average_mileage']= mpg[['city', 'hwy']].mean(axis=1)
mpg
```

```python
# Create a new column on the mpg dataset named is_automatic that holds boolean values denoting whether 
# the car has an automatic transmission.

mpg['is_automatic']= np.where(mpg['trans'].str.contains('auto'),'automatic','manual')
mpg
```

```python
# Using the mpg dataset, find out which which manufacturer has the best miles per gallon on average?

mpg.groupby('manufacturer').average_mileage.mean().sort_values().tail(1)

# honda, 28.5mpgs
```

```python
# Do automatic or manual cars have better miles per gallon?

mpg.groupby('is_automatic').average_mileage.mean()

# manual
```

```python
# Join the employees and titles DataFrames together.

url = f'mysql+pymysql://{user}:{password}@{host}/employees' 

sql = '''
SELECT
    *
FROM employees
'''

employees = pd.read_sql(sql, url)
employees.head()
```

```python
sql = '''
SELECT
    *
FROM titles
'''

titles = pd.read_sql(sql, url)
titles.head()
```

```python
emp_and_titles = employees.merge(titles, left_on='emp_no', right_on='emp_no', how='inner')
emp_and_titles
```

```python
# For each title, find the hire date of the employee that was hired most recently with that title.

emp_and_titles.groupby(['title']).hire_date.max()
```

```python
sql = '''
SELECT *
FROM departments;
'''

departments = pd.read_sql(sql, url)
departments
```

```python
sql = '''
SELECT *
FROM dept_emp;
'''

dept_emp = pd.read_sql(sql, url)
dept_emp
```

```python
dept_info = dept_emp.merge(departments, left_on='dept_no', right_on='dept_no', how='left')
```

```python
all_tables = emp_and_titles.merge(dept_info, left_on='emp_no', right_on='emp_no', how='outer')
```
