```python
import matplotlib.pyplot as plt  
import numpy as np
import math
from random import randint
import random
import pandas as pd
```

```python
# Use matplotlib to plot the following equation:

# y = x**2 − x + 2
# You'll need to write the code that generates the x and y points.

x = range(-10,10)
y = [(x**2) - x + 2 for x in x]

plt.plot(x,y)

plt.annotate('The Origin', xy= (0, 0), xytext = (.5, 25), arrowprops = dict(facecolor = 'purple')) 

# Add an anotation for the point 0, 0, the origin.
```

```python
# Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

# y = √x

x1 = range(0, 8)
y1 = np.sqrt(x1)

plt.plot(x1,y1)
```

```python
# y = x**3

x2 = range(0,10)
y2 = [n**3 for n in x2]

plt.plot(x2,y2)
```

```python
# y = 2**x

x3 = [x3 * 0.1 for x3 in range(0,50)]
y3 = [2 ** n for n in x3]

plt.plot(x3,y3)
```

```python
# y = 1/(x + 1)

x4 = [x4 * 0.1 for x4 in range(0,50)]
y4 = [1/(n+1) for n in x4]

plt.plot(x4,y4)
```

```python
# Combine the figures you created in the last step into one large figure with 4 subplots.


plt.figure(figsize = (12, 8))

plt.subplot(221)
plt.plot(x1,y1)
plt.subplot(222)
plt.plot(x2,y2)
plt.subplot(223)
plt.plot(x3,y3)
plt.subplot(224)
plt.plot(x4,y4)
```

```python
# Combine the figures you created in the last step into one figure where each of the 4 equations 
# has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

plt.figure(figsize = (12,8))

plt.plot(x1, y1, label='y = √x')
plt.plot(x2, y2, label='y = x**3')
plt.plot(x3, y3, label= 'y = 2**x')
plt.plot(x4, y4, label='y = 1/(x + 1)')

plt.ylim(0, 6)
plt.xlim(0, 2)

plt.legend()
plt.show()
```
