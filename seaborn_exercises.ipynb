```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pydataset import data
```

```python
# Use the iris database to answer the following quesitons:

iris = data('iris')
iris
```

```python
# What does the distribution of petal lengths look like?
sns.displot(data = iris, x = 'Petal.Length')

# the most frequent petal length is between 1-1.75
```

```python
# Is there a correlation between petal length and petal width? Use http://guessthecorrelation.com/ 
# as a hint to how we could visually determine if a correlation exists between two numeric columns.

sns.catplot(data = iris, x= 'Petal.Length', y = 'Petal.Width')
# yes, there is a correlation
```

```python
# Would it be reasonable to predict species based on sepal width and sepal length? 
# For this, you'll visualize two numeric columns through the lense of a categorical column.

sns.scatterplot(data = iris, x = 'Species', y='Sepal.Width')
sns.scatterplot(data = iris, x = 'Species', y='Sepal.Length')

# sepal length and width are not good indications of species
```

```python
# Which features would be best used to predict species?

sns.pairplot(iris, hue='Species')

#petal width and perhaps petal length would be the best indicators of species
```

```python
# Load the swiss dataset and read it's documentation. 
# The swiss dataset is available from pydatset rather than seaborn. 
# Create visualizations to answer the following questions:

swiss = data('swiss')
data('swiss', show_doc = True)
```

```python
swiss.info()
```

```python
# Create an attribute named is_catholic that holds a boolean value of whether or not the province is Catholic. 
# (Choose a cutoff point for what constitutes catholic)

is_catholic = swiss.Catholic > 50.00
is_catholic
```

```python
# Does whether or not a province is Catholic influence fertility?
sns.scatterplot(data=swiss, x=is_catholic, y='Fertility', hue= is_catholic)

#does not seem to have a large influence, I will try once more with a higher cutoff point for catholic
```

```python
is_catholic2 = swiss.Catholic > 80.00
is_catholic2

sns.scatterplot(data=swiss, x=is_catholic2, y='Fertility', hue= is_catholic2)

#with the cutoff for catholic being 80%, there is a slight positive correlation with fertility
```

```python
# What measure correlates most strongly with fertility?

sns.pairplot(swiss)

# I would say Agriculutre. Infant mortality looks like it may have noteable correlation with fertility as well
```

```python
# Using the chipotle dataset from the previous exercise, 
# create a bar chart that shows the 4 most popular items and the revenue produced by each.

url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
    
chipotle = pd.read_csv(url, sep = '\t')
chipotle.info()
```

```python
chipotle.item_name.value_counts().head(4)
```

```python
chipotle['item_price'] = chipotle.item_price.str.replace('$','').astype('float')
```

```python
sns.barplot(data=chipotle, x= chipotle.item_name.value_counts().head(4), 
                               y= 'item_price')
```

```python
Chicken_Bowl_Revenue = chipotle[chipotle['item_name']=='Chicken Bowl'].item_price.sum()
Chicken_Burrito_Revenue = chipotle[chipotle['item_name']=='Chicken Burrito'].item_price.sum()
Chips_and_Guacamole_Revenue = chipotle[chipotle['item_name']=='Chips and Guacamole'].item_price.sum()
Steak_Burrito_Revenue = chipotle[chipotle['item_name']=='Steak Burrito'].item_price.sum()
```
